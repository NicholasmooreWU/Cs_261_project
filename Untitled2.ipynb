```python
lineup_path = "/content/lineup_2024_USA.1.csv"  # Replace with the actual path to lineup.csv
lineup_df = pd.read_csv(lineup_path)
```

```python
import pandas as pd

# Load the dataset
df = lineup_df  # Assuming lineup_df is already loaded
df.replace(',,,', pd.NA, inplace=True)

# Create a new DataFrame to store formation and win information
formation_wins = []

# Iterate through each unique eventId
for event_id in df['eventId'].unique():
    # Get data for the current event
    event_data = df[df['eventId'] == event_id]

    # Extract formation and winner for home team
    formation_wins.append({
        'eventId': event_id,
        'formation': event_data[event_data['homeAway'] == 'home']['formation'].values[0],
        'win': 1 if event_data[event_data['homeAway'] == 'home']['winner'].values[0] == 1 else 0,
        'tie': 1 if event_data[event_data['homeAway'] == 'home']['winner'].values[0] == 0 and event_data[event_data['homeAway'] == 'away']['winner'].values[0] == 0 else 0,
        'lose': 1 if event_data[event_data['homeAway'] == 'home']['winner'].values[0] == 0 and event_data[event_data['homeAway'] == 'away']['winner'].values[0] == 1 else 0
    })

    # Extract formation and winner for away team
    formation_wins.append({
        'eventId': event_id,
        'formation': event_data[event_data['homeAway'] == 'away']['formation'].values[0],
        'win': 1 if event_data[event_data['homeAway'] == 'away']['winner'].values[0] == 1 else 0,
        'tie': 1 if event_data[event_data['homeAway'] == 'home']['winner'].values[0] == 0 and event_data[event_data['homeAway'] == 'away']['winner'].values[0] == 0 else 0,
        'lose': 1 if event_data[event_data['homeAway'] == 'away']['winner'].values[0] == 0 and event_data[event_data['homeAway'] == 'home']['winner'].values[0] == 1 else 0
    })

formation_wins_df = pd.DataFrame(formation_wins)

# Group by formation and calculate win percentage and count
formation_stats = formation_wins_df.groupby('formation').agg(
    total_count=('formation', 'size'),
    win_count=('win', 'sum'),
    tie_count=('tie', 'sum'),
    lose_count=('lose', 'sum')
).reset_index()

# Calculate win percentage
formation_stats['win_percentage'] = (formation_stats['win_count'] / formation_stats['total_count']) * 100
formation_stats['tie_percentage'] = (formation_stats['tie_count'] / formation_stats['total_count']) * 100
formation_stats['lose_percentage'] = (formation_stats['lose_count'] / formation_stats['total_count']) * 100

# Sort by win percentage in descending order
formation_stats = formation_stats.sort_values(by='win_percentage', ascending=False)

# Display the results
print(formation_stats)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


formation_stats = formation_stats.sort_values(by='total_count', ascending=False)

plt.figure(figsize=(12, 6))

sns.barplot(x='formation', y='win_percentage', data=formation_stats, label='Win', color='#4CAF50')
sns.barplot(x='formation', y='tie_percentage', data=formation_stats, label='Tie', bottom=formation_stats['win_percentage'], color='#2196F3')
sns.barplot(x='formation', y='lose_percentage', data=formation_stats, label='Lose', bottom=formation_stats['win_percentage'] + formation_stats['tie_percentage'], color='#F44336')


plt.title('Formation Win-Tie-Loss Rate (Stacked 100%)')
plt.xlabel('Formation')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Match Outcome', bbox_to_anchor=(1.05, 1), loc='upper left') # Moved legend outside
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the key events data
key_event_path = "/content/keyEvents_2024_ENG.1.csv"
key_event_df = pd.read_csv(key_event_path)

# Remove duplicate rows based on 'keyEventText' and 'updateDateTime'
key_event_df.drop_duplicates(subset=['keyEventText', 'updateDateTime'], inplace=True)

# Filter for yellow card events
yellow_cards_df = key_event_df[key_event_df['keyEventText'].str.contains('yellow card', case=False, na=False)]

# Calculate yellow cards per game
yellow_cards_per_game = yellow_cards_df.groupby('eventId')['keyEventText'].count().reset_index(name='yellow_card_count')

# Filter for games with more than 3 yellow cards
filtered_games = yellow_cards_per_game[yellow_cards_per_game['yellow_card_count'] > 3]

# Add a 'game_type' column to indicate playoff games
filtered_games['game_type'] = filtered_games['eventId'].apply(lambda x: 'Playoff' if x > 700000 else 'Regular Season')

# Create the bar plot, differentiating playoff games
plt.figure(figsize=(10, 6))
sns.barplot(x='eventId', y='yellow_card_count', hue='game_type', data=filtered_games, dodge=False)
plt.xlabel('Game ID')
plt.ylabel('Number of Yellow Cards')
plt.title('Yellow Cards per Game (More than 3 Yellow Cards)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.legend(title='Game Type')
plt.show()
```

```python
# prompt: get the win-lose-tie rate for each teamid in the lineup_2024_usa.1.csv
team_stats = []
# Calculate win-loss-tie rates for each teamId
for event_id in df['eventId'].unique():
    # Get data for the current event
    event_data = df[df['eventId'] == event_id]

    # Extract formation and winner for home team
    team_stats.append({
        'eventId': event_id,
        'teamId': event_data[event_data['homeAway'] == 'home']['teamId'].values[0],
        'win': 1 if event_data[event_data['homeAway'] == 'home']['winner'].values[0] == 1 else 0,
        'tie': 1 if event_data[event_data['homeAway'] == 'home']['winner'].values[0] == 0 and event_data[event_data['homeAway'] == 'away']['winner'].values[0] == 0  else 0,
        'lose': 1 if event_data[event_data['homeAway'] == 'home']['winner'].values[0] == 0 and event_data[event_data['homeAway'] == 'away']['winner'].values[0] == 1 else 0,
        #'none': 1 if event_data[event_data['homeAway'] == 'home']['winner'].values[0] == "" and event_data[event_data['homeAway'] == 'away']['winner'].values[0] == "" else
    })

    # Extract formation and winner for away team
    team_stats.append({
        'eventId': event_id,
        'teamId': event_data[event_data['homeAway'] == 'away']['teamId'].values[0],
        'win': 1 if event_data[event_data['homeAway'] == 'away']['winner'].values[0] == 1 else 0,
        'tie': 1 if event_data[event_data['homeAway'] == 'away']['winner'].values[0] == 0 and event_data[event_data['homeAway'] == 'home']['winner'].values[0] == 0  else 0,
        'lose': 1 if event_data[event_data['homeAway'] == 'away']['winner'].values[0] == 0 and event_data[event_data['homeAway'] == 'home']['winner'].values[0] == 1 else 0
    })
team_stats_df = pd.DataFrame(team_stats)
team_stats_df.dropna(inplace=True)

team_stats = team_stats_df.groupby('teamId').agg(
    wins=('win', 'sum'),
    losses=('lose', 'sum'),
    ties=('tie', 'sum'),
    total_matches=('win', 'count')
    ).reset_index()
team_stats['win_rate'] = (team_stats['wins'] / team_stats['total_matches']) * 100
team_stats['loss_rate'] = (team_stats['losses'] / team_stats['total_matches']) * 100
team_stats['tie_rate'] = (team_stats['ties'] / team_stats['total_matches']) * 100

print(team_stats[['teamId', 'total_matches', 'wins', 'losses', 'ties', 'win_rate', 'loss_rate', 'tie_rate']])
```
